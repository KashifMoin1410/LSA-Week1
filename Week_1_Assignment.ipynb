## Question 2: Given a list representing null counts per column: null_counts = [0, 2, 15, 1, 0, 8]
Write a program that:
* Iterates through the list
* Prints a warning for any column with null count > 5
* Prints total number of problematic columns

```python
def checkNullCounts(null_counts): # Function to check any column with null count > 5
    problematic_columns = 0 # Initial value set to 0

    for count in null_counts: # Iterate through the list
        if count > 5:
            print(f"Warning: Column has {count} null values")
            problematic_columns += 1 # Total count of problematic columns

    print("\nTotal number of problematic columns:", problematic_columns)


null_counts = [0, 2, 15, 1, 0, 8]
checkNullCounts(null_counts) # Function call passing the list.
```

## Question 3: Create a program that:
1. Stores a list of numeric values (e.g. speeds or emissions)
2. Uses a function to check if each value is within an acceptable range
3. Loops through the list and prints:
* Value
* Validation result
4.  Prints a final summary:
* Total values checked
* Total invalid values

```python
# Function to check if the speed is valid or not
def speedValidationFunction(value, min_value=0, max_value=100): # Since, not range for given, default values of 0 and 100 have been set.
    if min_value <= value <= max_value:
        return "Valid"
    else:
        return "Invalid"


speed = [random.randint(-20, 150) for _ in range(10)] # Randomly generate the list of integers

total_checked = 0
invalid_count = 0

print("Speed Validation Result:")
for value in speed:
    result = speedValidationFunction(value) # Function call to check of the speed is valid or not.
    print(f"Speed: {value} -> {result}")

    total_checked += 1 # Variable to store the count of total values checked
    if result == "Invalid":
        invalid_count += 1 # Toal invalid count.

print("\nFinal Summary:") # Print the final summary
print("Total values checked:", total_checked)
print("Total invalid values:", invalid_count)
```

```python
# Histogram to visualise the distribution of MPG
plt.figure(figsize=(8,5)) # Set the Figure size
plt.hist(df['MPG'], bins=20) # Histogram with 20 bars
plt.xlabel("Miles Per Gallon (MPG)") # Label for the x-axis
plt.ylabel("Frequency") # Label for the y-axis
plt.title("Distribution of MPG") # Plot title
plt.show()
```

### Data Visualisation

```python
# Box plot: MPG
plt.figure(figsize=(8,5)) # Set the Figure size
sns.boxplot(y='MPG', data=df)
plt.title("Box Plot of MPG") # Plot title
plt.ylabel("MPG") # Label for the y-axis
plt.show()
```

```python
# Scatter plot: Horsepower vs MPG
plt.figure(figsize=(8,5)) # Set the Figure size
sns.scatterplot(x='HP', y='MPG', data=df) # Scatter plot beetween MPG and HP
plt.xlabel("Horsepower") # Label for the x-axis
plt.ylabel("MPG") # Label for the y-axis
plt.title("Horsepower vs MPG") # Plot title
plt.show()
```
